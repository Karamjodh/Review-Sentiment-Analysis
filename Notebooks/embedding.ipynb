```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "the glass of milk",
    "the glass of juice",
    'I am a good boy',
    "I am a good developer",
    "understand the meaning of words",
    "your videos are good",
]
```

```python
sent
```

```python
voc_size = 10000
```

```python
one_hot_repr = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_repr
```

```python
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding = 'post',maxlen = sent_length)
```

```python
embedded_docs
```

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length = sent_length))
model.compile("adam","mse")
```

```python
model.summary()
```

Embedding layer are more preferable in Neural Networks because unlike the normal ML technique based vectorizers like TF-IDF, CBOW, Word2Vec Embedding layers are trainable generate vectors and through forward and backward propogation they keep on updating vectors according to the best context and semantic capture with in the model.
Emedding layers donot take direct text as input but text indices which can be generatd by tokenizer.text_to_sequence(words)

```python
model.predict(embedded_docs)
```
